````python
#your code
names=[]
for e in sopa.find_all('div', class_='col-md-6'):
    a=e.text.split('\n \n\n\n             ')            
    if len(a)>1:
        b=a[0].strip()
        c=a[1].replace(' ','(').replace('\n(\n',')')
        d=b+' '+c
              
    else:
        d= a[0].strip()+ ' ()'
    
    names.append(d)
    
names=names[::2]
names
````

####  20 latest earthquakes info (date, time, latitude, longitude and region name) by the EMSC as a pandas dataframe

````python
# This is the url you will scrape in this exercise
url = 'https://www.emsc-csem.org/Earthquake/'
````

````python
driver.get(url)
````

````python
driver.get(url)
xpath='/html/body/div[3]/div[6]/table'
tabla = driver.find_elements(By.XPATH, xpath)
a=tabla[0].text.split('[+]')[0].split('\n')

a.append('[+]')
a.append(tabla[0].text.split('[+]')[1].strip())
cabeceras=a[::2]
cabeceras.insert(-2,'nada')
cabeceras
````

````python
tabla = driver.find_element(By.XPATH, '/html/body/div[3]/div[6]/table/tbody')
filas = tabla.find_elements(By.TAG_NAME, 'tr')
data = []


for f in filas:
    
    elementos = f.find_elements(By.TAG_NAME, 'td')
    
    tmp_lst = []
    
    for e in elementos:
        #if e.text !='':
            tmp_lst.append(e.text)
        
    data.append(tmp_lst[2:])


data[:5]
````

````python
df=pd.DataFrame(data, columns=cabeceras)
df.drop('nada', axis=1)
df.head(10)
````

#### Top 10 languages by number of native speakers stored in a Pandas Dataframe

````python
#your code
html=req.get(url).text
html
sopa = bs(html, 'html.parser')

tabla=sopa.find_all('table')[0]
filas=tabla.find_all('tr')

tablacompleta=[]

for f in filas:
    tablacompleta.append(f.text.strip().split('\n\n'))
    
    
cabecera=tablacompleta[0]
tabla=tablacompleta[1:]
df=pd.DataFrame(tabla, columns=cabecera)
df.head(10)
````

#### IMDB's Top 250 data (movie name, Initial release, director name and stars) as a pandas dataframe

````python
# This is the url you will scrape in this exercise 
url = 'https://www.imdb.com/chart/top'
driver.get(url)
selector = '#__next > main > div > div.ipc-page-content-container.ipc-page-content-container--center > section > div > div.ipc-page-grid.ipc-page-grid--bias-left > div > ul > li:nth-child(1) > div.sc-e5a25b0f-0.jQjDIb.cli-poster-container > div > a > div'
#peliculas=driver.find_element(By.CSS_SELECTOR, selector).click()
tablas=driver.find_element(By.XPATH, '//*[@id="__next"]/main/div/div[3]/section/div/div[2]/div/ul')
peliculas=tablas.find_elements(By.TAG_NAME, 'li')
#[0].text.split('\n')[:2]
#By.TAG_NAME, 'tr'
a=peliculas[0].find_element(By.TAG_NAME, 'a')
a.get_attribute('href')
````

````python
links=[]
for p in peliculas: 
    a=p.find_element(By.TAG_NAME, 'a')
    b=a.get_attribute('href')
    links.append(b)
links
````

````python
import time

pelis=[]
for link in links: 
    try:
        driver.get(link)
        time.sleep(3)
        name=driver.find_element(By.XPATH, '//*[@id="__next"]/main/div/section[1]/section/div[3]/section/section/div[2]/div[1]/h1/span').text
        director=driver.find_element(By.XPATH, '//*[@id="__next"]/main/div/section[1]/div/section/div/div[1]/section[4]/ul/li[1]/div/ul/li/a').text
        release=driver.find_element(By.XPATH, '//*[@id="__next"]/main/div/section[1]/section/div[3]/section/section/div[2]/div[1]/ul/li[1]/a').text
        stars=driver.find_element(By.XPATH,'//*[@id="__next"]/main/div/section[1]/div/section/div/div[1]/section[4]/div[2]/div[2]').text.split('\n')[::2][:3]
        dictio={'name':name,'release':release,'director':director,'stars':stars}
        pelis.append(dictio)
    
    except:
        driver.get(link)
        time.sleep(3)
        name=driver.find_element(By.XPATH, '//*[@id="__next"]/main/div/section[1]/section/div[3]/section/section/div[2]/div[1]/h1/span').text
        release=driver.find_element(By.XPATH, '//*[@id="__next"]/main/div/section[1]/section/div[3]/section/section/div[2]/div[1]/ul/li[1]/a').textstars=driver.find_element(By.XPATH,'//*[@id="__next"]/main/div/section[1]/div/section/div/div[1]/section[4]/div[2]/div[2]').text
        director2=driver.find_element(By.XPATH, '//*[@id="__next"]/main/div/section[1]/div/section/div/div[1]/section[3]/ul/li[1]/div/ul/li/a').text
        stars2=driver.find_element(By.XPATH,'//*[@id="__next"]/main/div/section[1]/div/section/div/div[1]/section[3]/div[2]').text.split('\n')[::2][:3]
    

        dictio={'name':name,'release':release,'director':director2,'stars':stars2}
        pelis.append(dictio)

pelis
````

````python
df=pd.DataFrame(pelis)
pd.options.display.max_rows=None
df
````

#### Book name,price and stock availability as a pandas dataframe.

````python
# This is the url you will scrape in this exercise. 
# It is a fictional bookstore created to be scraped. 
url = 'http://books.toscrape.com/'
driver.get(url)
bloque=driver.find_element(By.XPATH, '//*[@id="default"]/div/div/div/div/section/div[2]/ol')
#nombre=libros[0].find_element(By.TAG_NAME, 'h3').find_element(By.TAG_NAME, 'a').get_attribute('title')
#precio=libros[0].find_element(By.XPATH, '//*[@id="default"]/div/div/div/div/section/div[2]/ol/li[1]/article/div[2]/p[1]').text.split('£')[1]
#stock=libros[0].find_element(By.XPATH, '//*[@id="default"]/div/div/div/div/section/div[2]/ol/li[1]/article/div[2]/p[2]').text

libros=bloque.find_elements(By.TAG_NAME, 'li')
````

````python
#your code
info=[]
for l in libros:

    nombre=l.find_element(By.TAG_NAME, 'h3').find_element(By.TAG_NAME, 'a').get_attribute('title')
    precio=l.find_element(By.XPATH, '//*[@id="default"]/div/div/div/div/section/div[2]/ol/li[1]/article/div[2]/p[1]').text.split('£')[1]
    stock=l.find_element(By.XPATH, '//*[@id="default"]/div/div/div/div/section/div[2]/ol/li[1]/article/div[2]/p[2]').text
    dictio={'nombre':nombre, 'precio': precio, 'stock':stock}
    info.append(dictio)
    
info
````

````python
df=pd.DataFrame(info)
df
````
