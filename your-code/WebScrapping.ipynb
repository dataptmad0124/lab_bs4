####  20 latest earthquakes info (date, time, latitude, longitude and region name) by the EMSC as a pandas dataframe

````python
# This is the url you will scrape in this exercise
url = 'https://www.emsc-csem.org/Earthquake/'
````

````python
#your code
browser.get(url)

tabla = browser.find_element(By.TAG_NAME, "table")
filas = tabla.find_elements(By.TAG_NAME, "tr")

data = []

for f in filas:
    elementos = f.find_elements(By.TAG_NAME, "td")
    
    temp_list = []
    
    for e in elementos:
        temp_list.append(e.text)
        
    data.append(temp_list)
    
cabeceras = tabla.find_element(By.XPATH, "/html/body/div[3]/div[6]/table/thead").text.split("\n")
cabeceras[0:2] = [''.join(cabeceras[0 : 2])]
````

````python
import pandas as pd

pd.DataFrame(data, columns=cabeceras)
````

#### Book name,price and stock availability as a pandas dataframe.

````python
# This is the url you will scrape in this exercise. 
# It is a fictional bookstore created to be scraped. 
url = 'http://books.toscrape.com/'
````

````python
#your code
browser.get(url)


book = browser.find_elements(By.TAG_NAME, "h3")
price = browser.find_elements(By.CLASS_NAME, "price_color")
stock = browser.find_elements(By.CLASS_NAME, "instock availability")

for b in book:
    names = (b.text)
    print(names)
    
for p in price:
    prices = p.text
    print(prices)
    
for s in stock:
    stockes = s.text
    print(stockes)

pd.DataFrame({names,prices,stockes})
````
